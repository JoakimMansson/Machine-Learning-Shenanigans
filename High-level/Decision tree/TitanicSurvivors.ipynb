```python
import pandas as pd
```

```python
dataset = pd.read_csv('titanic.csv')
dataset.head()
```

```python
input_data = dataset.drop(['Survived'], axis="columns")
target_data = dataset['Survived']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_Pclass = LabelEncoder()
le_Sex = LabelEncoder()
le_Parch = LabelEncoder()
le_Fare = LabelEncoder()
le_Cabin = LabelEncoder()
le_Embarked = LabelEncoder()
```

```python
# For encoding categorical variables to values
input_data["Pclass"] = le_Pclass.fit_transform(input_data["Pclass"])
input_data["Sex"] = le_Sex.fit_transform(input_data["Sex"])
input_data["Parch"] = le_Sex.fit_transform(input_data["Parch"])
input_data["Fare"] = le_Sex.fit_transform(input_data["Fare"])
input_data["Cabin"] = le_Sex.fit_transform(input_data["Cabin"])
input_data["Embarked"] = le_Sex.fit_transform(input_data["Embarked"])
input_data.head()
```

```python
inputs_data_n = inputs_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis="columns")
target_data_n = target_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis="columns")
```
